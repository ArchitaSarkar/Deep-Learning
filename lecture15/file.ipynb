# Gradient Descent and Its Types

In machine learning and deep learning, Gradient Descent is the process of adjusting the model's weights to minimize the loss (error) and reach the best possible result, known as the global minima.

---

# 1. Gradient Descent

In standard Gradient Descent, the model considers the entire dataset (all \(N\) records) to calculate the loss and update the weights at each step.

While this provides a direct path toward the global minima, it is extremely resource-intensive, especially with very large datasets (e.g., millions of records), as it requires loading the entire dataset into memory.

---

# 2. Stochastic Gradient Descent (SGD)

In pure Stochastic Gradient Descent, the model updates the weights using only one data point at a time.

This makes the process much faster per step, but the path toward the minima becomes very erratic or noisy because it reacts to the fluctuations of individual data points rather than the overall trend of the dataset.

---

# 3. Mini-Batch Stochastic Gradient Descent

This is the most common technique used in modern neural networks (like ANN or CNN).

It bridges the gap between the two methods.

---

## How it Works

Instead of using:
- One record
- Or all records

It uses a batch of \(B\) data points, where \(B\) is smaller than the total \(N\).

---

## Benefits

- Balances computational efficiency with stability
- More memory-friendly than full Gradient Descent
- Significantly less noisy and erratic than pure SGD

---

# Key Comparisons

## Convergence Path

### Gradient Descent
Moves in a smooth, direct line toward the global minima.

### Mini-Batch SGD
Follows a somewhat zigzag or noisy path because it is based on a sample of the population rather than the whole dataset, but it eventually converges.

---

## Computational Cost

### Gradient Descent
- High computational cost
- Slow for large datasets

### Mini-Batch SGD
- More scalable
- Standard approach for deep learning applications

---

# Additional Note

Because Mini-Batch SGD introduces some noise into the convergence path, advanced techniques like SGD with Momentum are often used.

Momentum helps:
- Smooth out movement
- Reduce oscillations
- Reach optimal weights more efficiently
