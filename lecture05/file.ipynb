# Weights and Biases

## Weights (W1, W2, W3)

Weights are assigned to the connections between inputs and neurons.

### Purpose of Weights
- Determine the importance of input signals
- Influence the neuron's final output
- Help the network learn patterns during training

Each input is multiplied by its corresponding weight.

## Bias

A bias value is added to the weighted sum before applying the activation function.

### Purpose of Bias
- Helps shift the activation function
- Improves learning capability
- Allows better flexibility in predictions

# Activation Function

The activation function determines whether a neuron should activate and pass information forward.

## Sigmoid Function

The speaker uses the Sigmoid activation function as an example.

### Formula

:contentReference[oaicite:1]{index=1}

### Properties
- Maps values between 0 and 1
- Used for probability-based outputs
- Smooth and differentiable

### Activation Rule
- If output > 0.5 → neuron activates
- If output ≤ 0.5 → neuron does not activate

# Human Nervous System Analogy

The activation process is compared to the human nervous system.

### Example
When a person touches a hot object:
1. Sensory neurons detect heat
2. Signals are processed
3. The brain reacts instantly
4. The hand is withdrawn

Similarly, artificial neurons process signals and decide whether to activate.

# Forward Propagation

Forward propagation is the process of passing information from:
1. Input Layer
2. Hidden Layer(s)
3. Output Layer

The network processes the data step-by-step until a final prediction is generated.
