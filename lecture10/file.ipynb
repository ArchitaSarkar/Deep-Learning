# Vanishing Gradient Problem

The **Vanishing Gradient Problem** is a common issue in deep neural networks where gradients become extremely small during training. As a result, the early layers of the network learn very slowly or stop learning completely.

---

# Why It Happens

Neural networks are trained using:
- Forward Propagation
- Back Propagation

During back propagation, gradients are calculated and propagated backward from the output layer to earlier layers.

When activation functions such as **Sigmoid** or **Tanh** are used repeatedly in deep networks, the gradients become smaller and smaller after every layer multiplication.

Eventually, the gradient approaches zero.

# Effects of Vanishing Gradient

## Problems Caused
- Early layers stop learning
- Training becomes very slow
- Model accuracy decreases
- Deep networks become difficult to train

---

# Example

Suppose a deep neural network has 20 hidden layers.

During back propagation:
- Output layers receive larger gradients
- Earlier layers receive almost zero gradients

As a result:
- Final layers learn properly
- Initial layers fail to learn meaningful features

# Summary

The Vanishing Gradient Problem occurs when gradients become extremely small during back propagation in deep neural networks.

This leads to:
- Slow learning
- Poor training
- Difficulty in training deep models

Common solutions include:
- ReLU activation
- Batch normalization
- Better initialization
- Residual connections
