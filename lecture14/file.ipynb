# Advanced Activation Functions in Deep Learning

Activation functions help neural networks learn complex patterns by deciding how much information should pass through a neuron.

Different activation functions are used for different types of neural networks and problems.

---

# What Softmax Does

Softmax converts raw output values into probabilities.

### Properties
- Output values lie between 0 and 1
- Total probability sum equals 1

---

## Example

Suppose a model predicts:
- Cat → 0.7
- Dog → 0.2
- Bird → 0.1

The model predicts:
- Cat with highest probability

---

# Choosing the Right Activation Function

Different activation functions are used for different purposes.

| Activation Function | Main Purpose |
|---|---|
| ReLU | General hidden layers |
| Leaky ReLU | Prevent dead neurons |
| ELU | Zero-centered outputs |
| PReLU | Learnable negative slope |
| Swish | Deep neural networks |
| Softplus | Smooth ReLU alternative |
| Softmax | Multi-class classification |

---

# Summary

Advanced activation functions improve:
- Learning capability
- Gradient flow
- Model performance
- Stability of deep networks

Choosing the correct activation function depends on:
- Network architecture
- Type of problem
- Depth of the neural network
