# Exploding Gradient Problem

The **Exploding Gradient Problem** occurs when gradients become extremely large during back propagation in a neural network. This causes huge weight updates, making the model unstable during training.

---

# Why It Happens

During back propagation, gradients are repeatedly multiplied across layers.

If the derivative values are greater than 1, the gradient grows exponentially.

As the number of layers increases, the gradient becomes extremely large.

---

# Effects of Exploding Gradient

## Problems Caused
- Very large weight updates
- Unstable training process
- Loss function becomes extremely high
- Model may produce NaN values
- Training may completely fail

---

# Example Scenario

Suppose a deep neural network has many hidden layers.

During back propagation:
- Gradients increase rapidly
- Weight values become extremely large
- The network fails to converge properly

Instead of learning patterns, the model becomes unstable.

# Common Causes

## 1. Deep Neural Networks
Very deep networks can amplify gradients during back propagation.

---

## 2. Large Weight Initialization
Large initial weights can cause gradients to grow rapidly.

---

## 3. High Learning Rate
A very high learning rate may lead to unstable updates.

---

# Solutions to Exploding Gradient Problem

## 1. Gradient Clipping

Gradient clipping limits gradient values within a fixed range.

### Example
If gradient exceeds a threshold:
- Reduce it to the threshold value

This prevents extremely large updates.

---

## 2. Proper Weight Initialization
Using smaller and well-balanced initial weights stabilizes training.

---

## 3. Batch Normalization
Helps maintain stable activations and gradients.

---

## 4. Lower Learning Rate
Reducing the learning rate prevents massive weight updates.

---

## 5. Using ReLU Carefully
While ReLU helps with vanishing gradients, proper initialization is still important to avoid exploding gradients.

---

# Difference Between Vanishing and Exploding Gradient

| Problem | Gradient Value | Effect |
|---|---|---|
| Vanishing Gradient | Very Small | Slow or no learning |
| Exploding Gradient | Very Large | Unstable training |

---

# Summary

The Exploding Gradient Problem happens when gradients become excessively large during back propagation.

This leads to:
- Unstable learning
- Huge weight updates
- Training failure

Common solutions include:
- Gradient clipping
- Proper initialization
- Batch normalization
- Smaller learning rate
