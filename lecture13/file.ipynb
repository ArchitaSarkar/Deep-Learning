# How ReLU and Leaky ReLU Solve the Vanishing Gradient Problem

The **Vanishing Gradient Problem** occurs when gradients become extremely small during back propagation in deep neural networks.

As a result:
- Early layers learn very slowly
- Weight updates become tiny
- Training becomes difficult

Activation functions like **Sigmoid** and **Tanh** are the main causes of this problem.

---

Its derivative is always very small.

During back propagation:
- Small derivatives are multiplied repeatedly
- Gradients become closer to zero

This causes vanishing gradients.

# How ReLU Solves Vanishing Gradient

For positive values:
- ReLU derivative is 1
- Gradients remain large enough
- Information flows properly through deep layers

# Problem with ReLU: Dying ReLU

If input becomes negative:
- Output becomes 0
- Gradient also becomes 0

The neuron may stop learning permanently.

This is called the **Dying ReLU Problem**.

# How Leaky ReLU Solves Vanishing Gradient

For negative inputs:
- Output is not exactly zero
- Small gradients still flow backward

This prevents neurons from dying completely.

### Advantages
- Reduces vanishing gradients
- Prevents dead neurons
- Improves learning stability

# Simple Intuition

## Sigmoid
Acts like a narrow pipe:
- Information flow becomes weak

## ReLU
Acts like a wide open path for positive values:
- Strong gradient flow

## Leaky ReLU
Keeps the path slightly open even for negative values:
- Prevents complete blockage

---

# Summary

ReLU and Leaky ReLU help solve the Vanishing Gradient Problem by keeping gradients large enough during back propagation.

## ReLU
- Uses derivative = 1 for positive inputs
- Allows faster and deeper learning

## Leaky ReLU
- Allows small gradients for negative inputs
- Prevents dead neurons
- Improves training stability

These activation functions are widely used in modern deep learning models.
