# AdaDelta and RMSProp Optimizers

AdaDelta and RMSProp are optimization algorithms used in Deep Learning to improve the problems faced by AdaGrad.

They help maintain a proper learning rate during training so the model can continue learning efficiently.

---

# Problem with AdaGrad

AdaGrad uses adaptive learning rates for different weights.

It stores the sum of all previous squared gradients.

### Problem
Over time:
- The accumulated gradient value becomes very large
- The learning rate becomes extremely small

As a result:
- Weight updates almost stop
- The model learns very slowly
- Training may stop before reaching the best solution

---

# Solution: AdaDelta and RMSProp

AdaDelta and RMSProp solve this problem by controlling the growth of accumulated gradients.

Instead of storing the complete sum of all previous gradients, they use an **exponentially weighted moving average**.

This helps:
- Keep learning rates stable
- Prevent learning from stopping
- Improve convergence speed

---

# How RMSProp and AdaDelta Work

Instead of:
- Adding all squared gradients forever

They:
- Give more importance to recent gradients
- Gradually forget older gradients

This prevents the accumulated value from becoming too large.

# Role of Gamma (\(\gamma\))

Gamma controls how much past information is remembered.

### Common Value
- \(\gamma = 0.95\)

### Effect
- Larger gamma → remembers more past gradients
- Smaller gamma → focuses more on recent gradients

---

# Advantages of RMSProp and AdaDelta

## Benefits
- Prevent learning rate from becoming too small
- Faster convergence
- More stable training
- Works well for deep neural networks
- Handles noisy gradients effectively

---

# Simple Intuition

## AdaGrad
Keeps collecting all past gradients forever.

Result:
- Learning rate continuously shrinks

---

## RMSProp / AdaDelta
Keeps only recent important gradients.

Result:
- Learning rate remains balanced
- Model continues learning properly

---

# Conclusion

AdaDelta and RMSProp improve AdaGrad by restricting the accumulation of gradients.

They use exponentially weighted moving averages to:
- Forget old gradients
- Maintain stable learning rates
- Achieve faster and smoother convergence

Because of these advantages, RMSProp is widely used in Deep Learning models.
